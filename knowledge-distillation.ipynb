```python
import os
import shutil
import moviepy.editor as mp
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import tensorflow as tf
import tensorflow_hub as hub
from tensorflow.keras import models, layers, optimizers, regularizers
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping, TensorBoard, ReduceLROnPlateau
from skimage.transform import resize
import librosa

# Function to extract audio from videos
def extract_audio_from_videos(input_dir, output_dir):
    for root, dirs, files in os.walk(input_dir):
        for file in files:
            if file.endswith('.mp4'):
                video_path = os.path.join(root, file)
                output_category_dir = os.path.join(output_dir, os.path.basename(root))
                os.makedirs(output_category_dir, exist_ok=True)
                audio_path = os.path.join(output_category_dir, file.replace('.mp4', '.wav'))
                try:
                    video_clip = mp.VideoFileClip(video_path)
                    video_clip.audio.write_audiofile(audio_path)
                    video_clip.close()
                except Exception as e:
                    print(f"Error processing {video_path}: {e}")

# Function to load and preprocess audio
def load_and_preprocess_audio(file_path, target_duration=10, target_sr=44100):
    try:
        # Load audio file with original sampling rate
        audio, sr = librosa.load(file_path, sr=None)

        # Resample if needed
        if sr != target_sr:
            audio = librosa.resample(y=audio, orig_sr=sr, target_sr=target_sr)
            sr = target_sr

        # Ensure target duration
        target_length = int(sr * target_duration)
        if len(audio) < target_length:
            audio = np.pad(audio, (0, target_length - len(audio)))
        elif len(audio) > target_length:
            audio = audio[:target_length]

        # Run the model, check the output
        embeddings = vggish_model(audio)
        embeddings.shape.assert_is_compatible_with([None, 128])
        return embeddings
    except Exception as e:
        print(f"Error processing {file_path}: {e}")
        return None

# Directory paths
violence_dir = '/kaggle/input/violence-dataset/Real Life Violence Dataset/violence'
non_violence_dir = '/kaggle/input/violence-dataset/Real Life Violence Dataset/non_violence'
output_dir = '/kaggle/working/output'

# Extract audio from videos
extract_audio_from_videos(violence_dir, output_dir)
extract_audio_from_videos(non_violence_dir, output_dir)

# Load VGGish model from URL
vggish_model = hub.load('https://kaggle.com/models/google/vggish/frameworks/TensorFlow2/variations/vggish/versions/1')

# Feature extraction and data preparation
X_train = []
y_train = []

for category in os.listdir(output_dir):
    category_dir = os.path.join(output_dir, category)
    for audio_file in os.listdir(category_dir):
        file_path = os.path.join(category_dir, audio_file)
        embedding = load_and_preprocess_audio(file_path)
        if embedding is not None:
            X_train.append(embedding)
            label = 0 if category == 'violence' else 1
            y_train.append(label)

X_train = np.array(X_train)
y_train = np.array(y_train)

# Define the model architecture
num_classes = 2
model = models.Sequential([
    layers.Flatten(input_shape=(X_train.shape[1], X_train.shape[2])),
    layers.Dense(3200, activation='relu', kernel_regularizer=regularizers.l2(1e-6)),
    layers.Dense(2400, activation='relu', kernel_regularizer=regularizers.l2(1e-6)),
    layers.Dense(1000, activation='relu', kernel_regularizer=regularizers.l2(1e-6)),
    layers.Dense(num_classes, activation='softmax')
])

# Compile the model
model.compile(optimizer=optimizers.Adam(learning_rate=0.0001),
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Train-validation split
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.1, random_state=42)

# Callbacks for training
checkpoint_callback = ModelCheckpoint(filepath='model_checkpoint.keras', monitor='val_accuracy', save_best_only=True)
early_stopping_callback = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
tensorboard_callback = TensorBoard(log_dir='./logs')
reduce_lr_callback = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, min_lr=1e-7)

# Train the model
history = model.fit(X_train, y_train,
                    batch_size=4,
                    epochs=100,
                    validation_data=(X_val, y_val),
                    callbacks=[checkpoint_callback, early_stopping_callback, tensorboard_callback, reduce_lr_callback])

# Save the trained model
model.save('violence_detection_model.h5')

# Function to smooth the curves
def smooth_curve(points, factor=0.8):
    smoothed_points = []
    for point in points:
        if smoothed_points:
            previous = smoothed_points[-1]
            smoothed_points.append(previous * factor + point * (1 - factor))
        else:
            smoothed_points.append(point)
    return smoothed_points

# Plot training & validation accuracy values
plt.figure(figsize=(12, 6))
plt.plot(smooth_curve(history.history['accuracy']), label='Train Accuracy')
plt.plot(smooth_curve(history.history['val_accuracy']), label='Validation Accuracy')
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.show()

# Plot training & validation loss values
plt.figure(figsize=(12, 6))
plt.plot(smooth_curve(history.history['loss']), label='Train Loss')
plt.plot(smooth_curve(history.history['val_loss']), label='Validation Loss')
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.show()

# Clean up
shutil.rmtree(output_dir)

print("Model training complete and saved.")
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision.transforms as transforms
from torch.utils.data import DataLoader, Dataset, TensorDataset
import os
import cv2
import numpy as np
from tqdm import tqdm
from sklearn.model_selection import StratifiedShuffleSplit
import timm
import imgaug.augmenters as iaa
import matplotlib.pyplot as plt
from PIL import Image

# Constants
IMG_SIZE = 224
COLOR_CHANNELS = 3
TEMPERATURE = 3.0
ALPHA = 0.5
FRAME_SAMPLE_RATE = 10  # Sample one frame every 10 frames
BATCH_SIZE = 4
NUM_WORKERS = 4  # Adjust as needed

# Function to convert video to frames with sampling and augmentation
def video_to_frames(video, sample_rate=FRAME_SAMPLE_RATE):
    vidcap = cv2.VideoCapture(video)
    ImageFrames = []
    frame_count = 0
    augmented_count = 0
    while vidcap.isOpened():
        success, image = vidcap.read()
        if success:
            if frame_count % sample_rate == 0:
                flip = iaa.Fliplr(1.0)
                zoom = iaa.Affine(scale=1.3)
                random_brightness = iaa.Multiply((1, 1.3))
                rotate = iaa.Affine(rotate=(-25, 25))

                image_aug = flip(image=image)
                image_aug = random_brightness(image=image_aug)
                image_aug = zoom(image=image_aug)
                image_aug = rotate(image=image_aug)

                rgb_img = cv2.cvtColor(image_aug, cv2.COLOR_BGR2RGB)
                resized = cv2.resize(rgb_img, (IMG_SIZE, IMG_SIZE))
                ImageFrames.append(resized)
                augmented_count += 1  # Increment augmented frame counter
            frame_count += 1
        else:
            break
    vidcap.release()
    return ImageFrames, augmented_count

# Data directories
PROJECT_DIR = '/kaggle/input/violence-dataset/Real Life Violence Dataset'

# Load video data and labels
X_original = []
y_original = []
augmented_frame_count = 0
CLASSES = ["violence", "non_violence"]

for category in os.listdir(PROJECT_DIR):
    if category == '.ipynb_checkpoints':
        continue

    path = os.path.join(PROJECT_DIR, category)
    class_num = CLASSES.index(category)
    for i, video in enumerate(tqdm(os.listdir(path)[:450])):  # Use the first 450 videos
        frames, aug_count = video_to_frames(os.path.join(path, video))
        augmented_frame_count += aug_count
        for frame in frames:
            X_original.append(frame)
            y_original.append(class_num)

# Convert lists to NumPy arrays
X_original = np.array(X_original, dtype=np.uint8).reshape(-1, IMG_SIZE, IMG_SIZE, 3)
y_original = np.array(y_original, dtype=np.int64)

# Stratified split into training and testing sets
stratified_sample = StratifiedShuffleSplit(n_splits=2, test_size=0.3, random_state=73)

for train_index, test_index in stratified_sample.split(X_original, y_original):
    X_train, X_test = X_original[train_index], X_original[test_index]
    y_train, y_test = y_original[train_index], y_original[test_index]

# Normalize and reshape data for PyTorch
X_train_nn = X_train / 255.0
X_test_nn = X_test / 255.0

# Convert to PyTorch tensors
X_train_nn = torch.tensor(X_train_nn, dtype=torch.float32).permute(0, 3, 1, 2)
X_test_nn = torch.tensor(X_test_nn, dtype=torch.float32).permute(0, 3, 1, 2)
y_train_nn = torch.tensor(y_train, dtype=torch.long)
y_test_nn = torch.tensor(y_test, dtype=torch.long)

# Create DataLoader
train_dataset = TensorDataset(X_train_nn, y_train_nn)
test_dataset = TensorDataset(X_test_nn, y_test_nn)

train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True, num_workers=NUM_WORKERS, pin_memory=True)
test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False, num_workers=NUM_WORKERS, pin_memory=True)

# Model definitions
class VisionTransformer(nn.Module):
    def __init__(self, num_classes=2):
        super(VisionTransformer, self).__init__()
        self.vit_model = timm.create_model("vit_base_patch16_224", pretrained=True)
        self.vit_model.head = nn.Linear(self.vit_model.head.in_features, num_classes)

    def forward(self, x):
        return self.vit_model(x)

class MobileViT(nn.Module):
    def __init__(self, num_classes=2, dropout=0.5):
        super(MobileViT, self).__init__()
        self.mobilevit = timm.create_model('mobilevit_xxs', pretrained=True)
        self.global_pool = nn.AdaptiveAvgPool2d((1, 1))
        self.classifier = nn.Sequential(
            nn.Dropout(p=dropout),
            nn.Linear(self.mobilevit.num_features, num_classes)
        )

    def forward(self, x):
        x = self.mobilevit.forward_features(x)
        x = self.global_pool(x)
        x = x.view(x.size(0), -1)
        x = self.classifier(x)
        return x

# Device configuration
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Initialize models
teacher_model = VisionTransformer().to(device)
student_model = MobileViT(dropout=0.5).to(device)

# Put teacher model in evaluation mode
teacher_model.eval()

# Define loss functions and optimizer
criterion = nn.CrossEntropyLoss()
distillation_loss = nn.KLDivLoss(reduction='batchmean')
optimizer = torch.optim.Adam(student_model.parameters(), lr=0.00001, weight_decay=1e-4)

# Custom Learning Rate Scheduler
start_lr = 0.00001
min_lr = 0.00001
max_lr = 0.00005
rampup_epochs = 5
sustain_epochs = 0
exp_decay = 0.8

def lrfn(epoch):
    if epoch < rampup_epochs:
        return (max_lr - start_lr) / rampup_epochs * epoch + start_lr
    elif epoch < rampup_epochs + sustain_epochs:
        return max_lr
    else:
        return (max_lr - min_lr) * exp_decay ** (epoch - rampup_epochs - sustain_epochs) + min_lr

class CustomLRScheduler:
    def __init__(self, optimizer):
        self.optimizer = optimizer

    def step(self, epoch):
        lr = lrfn(epoch)
        for param_group in self.optimizer.param_groups:
            param_group['lr'] = lr

# Early Stopping
class EarlyStopping:
    def __init__(self, patience=3, min_delta=0.00075):
        self.patience = patience
        self.min_delta = min_delta
        self.best_loss = None
        self.counter = 0

    def __call__(self, val_loss):
        if self.best_loss is None:
            self.best_loss = val_loss
        elif val_loss < self.best_loss - self.min_delta:
            self.best_loss = val_loss
            self.counter = 0
        else:
            self.counter += 1
            if self.counter >= self.patience:
                print("Early stopping triggered")
                return True
        return False

# Mixed Precision Training
scaler = torch.cuda.amp.GradScaler()

# Initialize lists to store loss and accuracy
train_losses = []
val_losses = []
train_accuracies = []
val_accuracies = []

# Training loop with early stopping and metric tracking
best_val_loss = float('inf')
early_stopping = EarlyStopping(patience=3)
scheduler = CustomLRScheduler(optimizer)
num_epochs = 100

# Gradient accumulation settings
accumulation_steps = 8  # Number of steps to accumulate gradients

for epoch in range(num_epochs):
    student_model.train()
    running_loss = 0.0
    correct_train = 0
    total_train = 0

    optimizer.zero_grad()

    try:
        for i, (frames, labels) in enumerate(tqdm(train_loader, desc=f"Epoch {epoch + 1}/{num_epochs}")):
            frames, labels = frames.to(device), labels.to(device)

            # Mixed precision training
            with torch.cuda.amp.autocast():
                outputs_student = student_model(frames)

                # Teacher model forward pass (detach to avoid gradient calculation)
                with torch.no_grad():
                    outputs_teacher = teacher_model(frames)

                # Compute losses
                loss_ce = criterion(outputs_student, labels)
                loss_kd = distillation_loss(
                    nn.functional.log_softmax(outputs_student / TEMPERATURE, dim=1),
                    nn.functional.softmax(outputs_teacher / TEMPERATURE, dim=1)
                ) * (TEMPERATURE ** 2)
                loss = ALPHA * loss_ce + (1 - ALPHA) * loss_kd

            scaler.scale(loss).backward()

            if (i + 1) % accumulation_steps == 0 or (i + 1) == len(train_loader):
                scaler.step(optimizer)
                scaler.update()
                optimizer.zero_grad()  # Move optimizer.zero_grad() inside the if condition

            running_loss += loss.item()

            # Compute training accuracy
            _, predicted_train = torch.max(outputs_student, 1)
            total_train += labels.size(0)
            correct_train += (predicted_train == labels).sum().item()

    except Exception as e:
        # Catch exceptions and log them
        print(f"Exception occurred in epoch {epoch + 1}: {str(e)}")
        continue

    epoch_loss = running_loss / len(train_loader)
    train_accuracy = correct_train / total_train
    train_losses.append(epoch_loss)
    train_accuracies.append(train_accuracy)
    print(f"Train Loss: {epoch_loss:.4f}, Train Accuracy: {train_accuracy:.4f}")

    # Validation loop
    student_model.eval()
    val_loss = 0.0
    correct_val = 0
    total_val = 0
    with torch.no_grad():
        try:
            for frames, labels in tqdm(test_loader, desc="Validation"):
                frames, labels = frames.to(device), labels.to(device)

                outputs_student = student_model(frames)
                _, predicted_val = torch.max(outputs_student, 1)

                total_val += labels.size(0)
                correct_val += (predicted_val == labels).sum().item()

                val_loss += criterion(outputs_student, labels).item()

        except Exception as e:
            # Catch exceptions and log them
            print(f"Exception occurred during validation in epoch {epoch + 1}: {str(e)}")
            continue

    val_loss /= len(test_loader)
    val_accuracy = correct_val / total_val
    val_losses.append(val_loss)
    val_accuracies.append(val_accuracy)
    print(f"Validation Loss: {val_loss:.4f}, Validation Accuracy: {val_accuracy:.4f}")

    # Check for early stopping
    if early_stopping(val_loss):
        break

    # Save the best model
    if val_loss < best_val_loss:
        best_val_loss = val_loss
        torch.save(student_model.state_dict(), 'best_model.pth')

    # Adjust learning rate
    scheduler.step(epoch)

# Load the best model
student_model.load_state_dict(torch.load('best_model.pth'))

# Evaluate on the test set
student_model.eval()
correct_test = 0
total_test = 0
test_predictions = []

with torch.no_grad():
    for frames, labels in tqdm(test_loader, desc="Testing"):
        frames, labels = frames.to(device), labels.to(device)

        outputs_student = student_model(frames)
        _, predicted_test = torch.max(outputs_student, 1)
        test_predictions.extend(predicted_test.cpu().numpy())

        total_test += labels.size(0)
        correct_test += (predicted_test == labels).sum().item()

test_accuracy = correct_test / total_test
print(f"Test Accuracy: {test_accuracy:.4f}")

# Save test predictions to a file
with open('test_predictions.txt', 'w') as f:
    for pred in test_predictions:
        f.write(f"{pred}\n")

# Plot training and validation loss
plt.figure(figsize=(12, 6))
plt.plot(train_losses, label='Training Loss')
plt.plot(val_losses, label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training and Validation Loss')
plt.legend()
plt.show()

# Plot training and validation accuracy
plt.figure(figsize=(12, 6))
plt.plot(train_accuracies, label='Training Accuracy')
plt.plot(val_accuracies, label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Training and Validation Accuracy')
plt.legend()
plt.show()

# Print the number of augmented frames
print(f"Total number of augmented frames: {augmented_frame_count}")
```

```python
import os
import shutil
import moviepy.editor as mp
import numpy as np
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
import tensorflow as tf
import tensorflow_hub as hub
from tensorflow.keras.models import load_model
from skimage.transform import resize
import librosa
import cv2
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, TensorDataset
import imgaug.augmenters as iaa
from tqdm import tqdm
import matplotlib.pyplot as plt
import seaborn as sns

# Device configuration
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Set CUDA_LAUNCH_BLOCKING for debugging
os.environ["CUDA_LAUNCH_BLOCKING"] = "1"

# Constants
IMG_SIZE = 224
FRAME_SAMPLE_RATE = 10
BATCH_SIZE = 4
NUM_WORKERS = 0  # Set to 0 for debugging

# Load the MobileViT model
student_model = MobileViT(dropout=0.5).to(device)
student_model.load_state_dict(torch.load('best_model.pth'))
student_model.eval()

# Load the audio-based model
audio_model = load_model('violence_detection_model.h5')

# Load VGGish model for audio feature extraction
vggish_model = hub.load('https://kaggle.com/models/google/vggish/frameworks/TensorFlow2/variations/vggish/versions/1')

# Function to extract frames from a video
def extract_frames_from_video(video_path, sample_rate=FRAME_SAMPLE_RATE):
    vidcap = cv2.VideoCapture(video_path)
    frames = []
    frame_count = 0
    while vidcap.isOpened():
        success, image = vidcap.read()
        if success:
            if frame_count % sample_rate == 0:
                flip = iaa.Fliplr(1.0)
                zoom = iaa.Affine(scale=1.3)
                random_brightness = iaa.Multiply((1, 1.3))
                rotate = iaa.Affine(rotate=(-25, 25))

                image_aug = flip(image=image)
                image_aug = random_brightness(image=image_aug)
                image_aug = zoom(image=image_aug)
                image_aug = rotate(image=image_aug)

                rgb_img = cv2.cvtColor(image_aug, cv2.COLOR_BGR2RGB)
                resized = cv2.resize(rgb_img, (IMG_SIZE, IMG_SIZE))
                frames.append(resized)
            frame_count += 1
        else:
            break
    vidcap.release()
    return frames if frames else None

# Function to extract audio from videos
def extract_audio_from_video(video_path, output_audio_path):
    try:
        video_clip = mp.VideoFileClip(video_path)
        if video_clip.audio is None:
            return None
        video_clip.audio.write_audiofile(output_audio_path)
        video_clip.close()
        return output_audio_path
    except Exception as e:
        print(f"Error processing {video_path}: {e}")
        return None

# Function to load and preprocess audio
def load_and_preprocess_audio(file_path, target_duration=10, target_sr=44100):
    # Load audio file with original sampling rate
    audio, sr = librosa.load(file_path, sr=None)

    # Resample if needed
    if sr != target_sr:
        audio = librosa.resample(y=audio, orig_sr=sr, target_sr=target_sr)
        sr = target_sr

    # Ensure target duration
    target_length = int(sr * target_duration)
    if len(audio) < target_length:
        audio = np.pad(audio, (0, target_length - len(audio)))
    elif len(audio) > target_length:
        audio = audio[:target_length]

    # Run the model, check the output
    embeddings = vggish_model(audio)
    embeddings.shape.assert_is_compatible_with([None, 128])
    return embeddings

# Function to get vision model predictions
def get_vision_predictions(model, video_paths):
    model.eval()
    video_predictions = {}
    with torch.no_grad():
        for video_path in tqdm(video_paths, desc="Vision Model Prediction"):
            frames = extract_frames_from_video(video_path)
            if frames is None:
                print(f"No frames extracted from {video_path}")
                continue

            frames = np.array(frames, dtype=np.uint8).reshape(-1, IMG_SIZE, IMG_SIZE, 3) / 255.0
            frames = torch.tensor(frames, dtype=torch.float32).permute(0, 3, 1, 2).to(device)
            loader = DataLoader(TensorDataset(frames), batch_size=BATCH_SIZE, shuffle=False, num_workers=NUM_WORKERS)

            video_preds = []
            for batch in loader:
                outputs = model(batch[0])
                probs = nn.functional.softmax(outputs, dim=1).cpu().numpy()
                video_preds.extend(probs)

            video_predictions[video_path] = np.mean(video_preds, axis=0)
    
    return video_predictions

# Function to get audio model predictions
def get_audio_predictions(model, video_paths, temp_audio_dir):
    predictions = {}
    for video_path in tqdm(video_paths, desc="Audio Model Prediction"):
        audio_output_path = os.path.join(temp_audio_dir, os.path.basename(video_path).replace('.mp4', '.wav'))
        audio_file = extract_audio_from_video(video_path, audio_output_path)
        if audio_file is None:
            print(f"No audio extracted from {video_path}")
            continue

        embedding = load_and_preprocess_audio(audio_file)
        if embedding is None:
            print(f"No embeddings generated for {video_path}")
            continue

        embedding = np.expand_dims(embedding, axis=0)
        prob = model.predict(embedding)
        predictions[video_path] = prob[0]

    return predictions

# Directory paths
test_videos_dir = '/kaggle/input/test-1/Test'
temp_audio_dir = '/kaggle/working/temp_audio'

# Create a writable directory for temporary audio files
os.makedirs(temp_audio_dir, exist_ok=True)

# List all test video files with labels
video_paths = []
true_labels = []

violence_videos = [os.path.join(test_videos_dir, 'violence', f) for f in os.listdir(os.path.join(test_videos_dir, 'violence')) if f.endswith('.mp4')]
non_violence_videos = [os.path.join(test_videos_dir, 'non_violence', f) for f in os.listdir(os.path.join(test_videos_dir, 'non_violence')) if f.endswith('.mp4')]

for video in violence_videos:
    video_paths.append(video)
    true_labels.append(0)  # Violence is labeled as 0

for video in non_violence_videos:
    video_paths.append(video)
    true_labels.append(1)  # Non-violence is labeled as 1

# Get vision model predictions
vision_predictions = get_vision_predictions(student_model, video_paths)

# Get audio model predictions
audio_predictions = get_audio_predictions(audio_model, video_paths, temp_audio_dir)

# Ensemble predictions
final_predictions = []
vision_weight = 0.6
audio_weight = 0.4

for video_path in video_paths:
    if video_path in vision_predictions and video_path in audio_predictions:
        ensemble_prediction = (vision_weight * vision_predictions[video_path] + audio_weight * audio_predictions[video_path])
        final_predictions.append(np.argmax(ensemble_prediction))
    elif video_path in vision_predictions:
        final_predictions.append(np.argmax(vision_predictions[video_path]))

# Calculate confusion matrix and accuracy
conf_matrix = confusion_matrix(true_labels, final_predictions)
accuracy = accuracy_score(true_labels, final_predictions)

print(f"Confusion Matrix:\n{conf_matrix}")
print(f"Accuracy: {accuracy:.4f}")

# Generate the classification report
class_report = classification_report(true_labels, final_predictions, target_names=['Violence', 'Non-violence'])

# Print the classification report
print("Classification Report:\n", class_report)

# Plot and save the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['Violence', 'Non-violence'], yticklabels=['Violence', 'Non-violence'])
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix')
plt.savefig('confusion_matrix.png')
plt.show()

# Save the confusion matrix and accuracy to a file
with open('evaluation_results.txt', 'w') as f:
    f.write(f"Confusion Matrix:\n{conf_matrix}\n")
    f.write(f"Accuracy: {accuracy:.4f}\n")

# Save the classification report to the file
with open('evaluation_results.txt', 'a') as f:
    f.write("\nClassification Report:\n")
    f.write(class_report)
```
